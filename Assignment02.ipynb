```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
```

**Performing** **Data** **Visualization**

1.UNIVARIATE ANALYSIS

```python
sns.distplot(df.speeding)
```

Inference : This is how the distribution in the speeding parameter using the distribution plot, We can see more density when speeding is 4.032.  

```python
sns.displot(df.alcohol)
```

Inference : we can see the displot layout of alcohol parameter, We can see the highest count when alcohol range is between 4 to 5.

```python
plt.pie(df.abbrev.value_counts())
plt.show()
```

Inference : We can see pie chart for each and every obseravation of abbrev parameter. We can see total 51 observations using the pie chart.

```python
df.abbrev.unique()
```

```python
df.abbrev.value_counts()
```

```python
sns.barplot(x=df.abbrev.value_counts().index, y=df.abbrev.value_counts())
```

Inference : We can see that all observations with same frequency in the abbrev parameter or feature or independent variable.


2.BIVARIATE ANALYSIS

```python
sns.lineplot(x=df.speeding,y=df.alcohol)
```

Inference : We can see complete irregular curve by lineplot between the speeding and alcohol.  

```python
sns.scatterplot(x=df.not_distracted,y=df.no_previous)
```

Inference : We can see complete irregular points by scatter plot between the not distracted and no previous.

```python
sns.jointplot(x=df.ins_premium,y=df.ins_losses)
```

Inference : We can see complete irregular points by scatter plot between the insurance premium and insurance loss. As well as here we can see individual bar graph of both insurance premium and insurance loss.

3.MULTIVARIATE ANALYSIS

```python
sns.pairplot(df)
```

Inference : Here we can see complete plots as pairs between each and every combination of all parameters.

```python
sns.heatmap(df.corr(),annot=True)
```

Inference : Clearly we can see there is no Multicollinearity and all Light colors indicate high correlation and Dark colors indicate low correlation.

```python
sns.boxplot(df.alcohol)
```

Inference : We can see very few ouliers beyond the upper limit.

```python
sns.boxplot(df)
```

Inference : We can see that using box plot we can detect outliers for a single parameter as well as for multiple parameters.
